```python
import sys
sys.path.append('..')

from osmuf.core import *
```

```python
# dictionary of places of interest for convenience, 
places = {'posadas' : (-27.37577,-55.90059),
          'kaduna' : (10.53364, 7.37360),
          'buenos_aires' : (-34.5798,-58.4422), #57897,44013
          'margery_street' : (51.5266,-0.1109),
          'welwyn_garden_city': (51.7938,-0.18645)}
```

```python
place='margery_street'
point=places[place]
distance=500 # in meters
```

```python
city_blocks_net, city_blocks_gross, city_blocks_gross_raw = city_blocks(point, distance)
```

```python
ax = city_blocks_gross_raw.plot(color='lightgrey', edgecolor='white', figsize=(16,16));

city_blocks_gross.plot(ax=ax, color='green',edgecolor='white', alpha=0.7);

city_blocks_net.plot(ax=ax, color='red', alpha=0.5);

# label style
style = dict(size=11, color='black', horizontalalignment='center')

# show areas of gross city blocks
for idx, row in city_blocks_net.iterrows():
    label = str(round((row.area_ha_net),2)) + '\n (' + str(round((row.net_to_gross),2)) + ')'
    ax.text(row.geometry.centroid.x, row.geometry.centroid.y, label, **style)
```
